```python
kidney_df = pd.read_csv("datasets/kidney_disease.csv")
```

```python
kidney_df.head()
```

```python
kidney_df.info()
```

```python
kidney_df.isnull().sum()
```

```python
kidney_df.replace("?", np.nan, inplace=True)
```

```python
kidney_df.dtypes
```

```python
numeric_columns = [
    "id", "age", "bp", "sg", "al", "su",
    "bgr", "bu", "sc", "sod", "pot", "hemo",
    "pcv", "wc", "rc"
]

for col in numeric_columns:
    kidney_df[col] = pd.to_numeric(kidney_df[col], errors="coerce")
```

```python
for col in numeric_columns:
    kidney_df[col] = kidney_df[col].fillna(kidney_df[col].median())
```

```python
categorical_columns = [
    "rbc", "pc", "pcc", "ba",
    "htn", "dm", "cad",
    "appet", "pe", "ane"
]

for col in categorical_columns:
    kidney_df[col] = kidney_df[col].fillna(kidney_df[col].mode()[0])
```

```python
kidney_df["classification"] = kidney_df["classification"].replace({
    "ckd": 1,
    "ckd\t": 1,
    "notckd": 0
})
```

```python
print(kidney_df["classification"].value_counts())
print(kidney_df["classification"].unique())
```

```python
from sklearn.preprocessing import LabelEncoder

encoder = LabelEncoder()

for col in categorical_columns:
    kidney_df[col] = encoder.fit_transform(kidney_df[col].astype(str))
```

```python
print(kidney_df["classification"].value_counts())
print(kidney_df["classification"].unique())
```

```python
kidney_df.isnull().sum()
```

```python
from sklearn.preprocessing import StandardScaler

X = kidney_df.drop("classification", axis=1)
y = kidney_df["classification"]

scaler = StandardScaler()

X_scaled = scaler.fit_transform(X)

processed_kidney = pd.DataFrame(X_scaled, columns=X.columns)
processed_kidney["classification"] = y
```

```python
processed_kidney.to_csv("datasets/processed_kidney.csv", index=False)
```

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from xgboost import XGBClassifier
from sklearn.metrics import (
    accuracy_score,
    precision_score,
    recall_score,
    f1_score,
    roc_auc_score
)
import joblib
```

```python
diabetes_dataset_df = pd.read_csv("datasets/processed_diabetes.csv")
```

```python
diabetes_dataset_df.head()
```

```python
X_diabetes = diabetes_dataset_df.drop("Outcome", axis=1)
y_diabetes = diabetes_dataset_df["Outcome"]
```

```python
X_train_diabetes, X_test_diabetes, y_train_diabetes, y_test_diabetes = train_test_split(
    X_diabetes,
    y_diabetes,
    test_size=0.2,
    random_state=42
)
```

```python
diabetes_model = XGBClassifier(
    random_state=42,
    eval_metric="logloss"
)
```

```python
diabetes_model.fit(X_train_diabetes, y_train_diabetes)
```

```python
predictions = diabetes_model.predict(X_test_diabetes)
probabilities = diabetes_model.predict_proba(X_test_diabetes)[:, 1]
```

```python
accuracy = accuracy_score(y_test_diabetes, predictions)
precision = precision_score(y_test_diabetes, predictions)
recall = recall_score(y_test_diabetes, predictions)
f1 = f1_score(y_test_diabetes, predictions)
roc_auc = roc_auc_score(y_test_diabetes, probabilities)

print(f"Accuracy : {accuracy * 100:.2f}%")
print(f"Precision: {precision * 100:.2f}%")
print(f"Recall   : {recall * 100:.2f}%")
print(f"F1 Score : {f1 * 100:.2f}%")
print(f"ROC-AUC  : {roc_auc * 100:.2f}%")
```

```python
joblib.dump(diabetes_model, "backend/ml/models/xgboost_diabetes_model.pkl")
```
