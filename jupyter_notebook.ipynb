```python
kidney_df = pd.read_csv("datasets/kidney_disease.csv")
```

```python
kidney_df.head()
```

```python
kidney_df.info()
```

```python
kidney_df.isnull().sum()
```

```python
kidney_df.replace("?", np.nan, inplace=True)
```

```python
kidney_df.dtypes
```

```python
numeric_columns = [
    "id", "age", "bp", "sg", "al", "su",
    "bgr", "bu", "sc", "sod", "pot", "hemo",
    "pcv", "wc", "rc"
]

for col in numeric_columns:
    kidney_df[col] = pd.to_numeric(kidney_df[col], errors="coerce")
```

```python
for col in numeric_columns:
    kidney_df[col] = kidney_df[col].fillna(kidney_df[col].median())
```

```python
categorical_columns = [
    "rbc", "pc", "pcc", "ba",
    "htn", "dm", "cad",
    "appet", "pe", "ane",
    "classification"
]

for col in categorical_columns:
    kidney_df[col] = kidney_df[col].fillna(kidney_df[col].mode()[0])
```

```python
from sklearn.preprocessing import LabelEncoder

encoder = LabelEncoder()

for col in categorical_columns:
    kidney_df[col] = encoder.fit_transform(kidney_df[col].astype(str))
```

```python
kidney_df.isnull().sum()
```

```python
from sklearn.preprocessing import StandardScaler

X = kidney_df.drop("classification", axis=1)
y = kidney_df["classification"]

scaler = StandardScaler()

X_scaled = scaler.fit_transform(X)

processed_kidney = pd.DataFrame(X_scaled, columns=X.columns)
processed_kidney["classification"] = y
```

```python
processed_kidney.to_csv("datasets/processed_kidney.csv", index=False)
```
